```python
import os,sys,inspect
from collections import defaultdict
from itertools import product
from scipy.io import wavfile
import importlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir)
```

```python
def create_dataframe(dataset, dataframe_dict):
    dir_path = os.path.join(PROJECT_ROOT_DIR, 'data', 'processed', 'preprocessed', dataset)
    for filename in os.listdir(dir_path):
        metadata = filename.split('_')
        dataframe_dict['fname'].append(filename)
        dataframe_dict['label'].append(metadata[0])
        dataframe_dict['dataset'].append(metadata[1])
        filepath = os.path.join(dir_path, filename)
        signal = wavfile.read(filepath)[1]
        for threshold in thresholds:
            dataframe_dict[str(threshold)].append(len(np.argwhere(np.diff(np.sign(threshold - signal))).flatten()))
    return dataframe_dict
```

```python
thresholds = np.arange(-6,7,1)
dataframe_dict = defaultdict(list)
dataframe_dict = create_dataframe('kaggle', dataframe_dict)
dataframe_dict = create_dataframe('physionet', dataframe_dict)
dataframe = pd.DataFrame(dataframe_dict)
```
