```python
f, ax = plt.subplots(figsize=(6, 6))
sns.barplot(y='sampling_frequency', x='set', data=physionet_overview, hue='label')
f.suptitle('Sampling frequency in each set by label', fontsize=18, fontweight='bold')
ax.set_xlabel("Label",size = 16,alpha=0.7)
ax.set_ylabel("Sampling frequency",size = 16,alpha=0.7)
plt.legend(loc='upper left')
```

```python
sns.set(style="whitegrid", palette="pastel", color_codes=True)

f, ax = plt.subplots(figsize=(8, 8))

sns.violinplot(x="set",y="magnitude_bandwidth",hue="label", data=physionet_overview)
sns.despine(left=True)

f.suptitle('Magnitude bandwidth by set in PhysioNet', fontsize=18, fontweight='bold')
ax.set_xlabel("Set",size = 16,alpha=0.7)
ax.set_ylabel("Bandwidth",size = 16,alpha=0.7)
plt.legend(loc='upper left')
```

```python
sns.set(style="whitegrid", palette="pastel", color_codes=True)

f, ax = plt.subplots(figsize=(8, 8))

sns.violinplot(x="set",y="recording_length",hue="label", data=physionet_overview)
sns.despine(left=True)

f.suptitle('Recording length by set in PhysioNet', fontsize=18, fontweight='bold')
ax.set_xlabel("Set",size = 16,alpha=0.7)
ax.set_ylabel("Recording length",size = 16,alpha=0.7)
plt.legend(loc='upper left')
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.countplot(x='set', data=physionet_overview, hue='label')
f.suptitle('Number of recordings in each set by label', fontsize=18, fontweight='bold')
ax.set_xlabel("Label",size = 16,alpha=0.7)
ax.set_ylabel("Number of recordings",size = 16,alpha=0.7)
plt.legend(loc='upper left')
```
