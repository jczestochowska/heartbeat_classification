```python
sns.set(style="whitegrid", palette="pastel", color_codes=True)

f, ax = plt.subplots(figsize=(8, 8))

plot = sns.violinplot(x="baza",y="recording_length",hue="etykieta", data=physionet_overview)
sns.despine(left=True)

f.suptitle('', fontsize=18, fontweight='bold')
ax.set_xlabel("Baza",size = 16,alpha=0.7)
ax.set_ylabel("Dugo nagrania",size = 16,alpha=0.7)
plt.legend(loc='upper left')
fig = plot.get_figure()
fig.savefig('recording_length.jpg')
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.countplot(x='set', data=physionet_overview, hue='label')
f.suptitle('Number of recordings in each set by label', fontsize=18, fontweight='bold')
ax.set_xlabel("Label",size = 16,alpha=0.7)
ax.set_ylabel("Number of recordings",size = 16,alpha=0.7)
plt.legend(loc='upper left')
```

```python
physionet_overview['etykieta'] = physionet_overview['label'].map({'abnormal': 'zaburzone', 'normal': 'zdrowe'})
physionet_overview['baza'] = physionet_overview['set'].map({'set_a': 'a', 'set_b': 'b', 'set_c':'c', 'set_d': 'd', 'set_e': 'e','set_f':'f'})
```
